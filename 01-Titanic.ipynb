## Titanic cleaning

### Dealing with missing values

```python
df.isna().sum()
```

```python
# removing rows
df.dropna(subset=["Embarked"], inplace=True) #inplace true is drop and also update df
```

```python
df.isna().sum()
```

```python
# removing columns
df.drop(columns=["Cabin"], inplace=True) 
```

```python
df.isna().sum()
```

```python
df.describe()
```

#### Replacing with 0

```python
df_new = df.copy()
```

```python
df_new.fillna({"Age": 0}, inplace=True)
```

```python
df_new.isna().sum()
```

```python
df_new.describe()
```

#### Replacing with mean

```python
df_new = df.copy()
```

```python
mean_age = df_new["Age"].mean()
mean_age
```

```python
df_new.fillna({"Age": mean_age}, inplace=True)
```

```python
df_new.isna().sum()
```

```python
df_new.describe()
```

#### Replacing with Previous Record

```python
df_new = df.copy()
```

```python
df_new.head()
```

```python
df_new["Age"].dtype
```

```python
df_new = df_new.astype({"Age": "float64"})
df_new.dtypes
```

```python
df_new["Age"] = df_new["Age"].ffill()
df_new.isna().sum()
```

```python
df_new.describe()
```

#### Replacing with Average fill (ffill + bfill)/2

```python
df_new = df.copy()
```

```python
df_new = df_new.astype({"Age": "float64"})
df_new.dtypes
```

```python
df_new["Age"] = df_new["Age"].ffill().add(df_new["Age"].bfill()).div(2)
df_new.isna().sum()
```

```python
df_new.describe()
```

## Converting Sex to Numbers

```python
df_copy = df.copy()

for index, row in df.iterrows():
    if row["Sex"] == "male":
        df_copy.loc[index,"Sex"] = 1
    else:
        df_copy.loc[index,"Sex"] = 0

df_copy.head(10)
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
q1 = np.quantile(df["Age"], 0.25)
q3 = np.quantile(df["Age"], 0.75)
print(f"Q1: {q1}, Q3: {q3}")

iqr = q3 - q1
print(f"IQR: {iqr}")

maximum = q3 + (1.5 * iqr)
minimum = q1 - (1.5 * iqr)
print(f"Min: {minimum}, Max: {maximum}")
```
