## Converting Sex to Numbers

```python
df_copy = df.copy()

t0 = time.time()

for index, row in df_copy.iterrows():
    if row["Sex"] == "male":
        df_copy.loc[index, "Sex"] = 1
    else:
        df_copy.loc[index, "Sex"] = 0

t1 = time.time()
print(f"Processed in {t1-t0:.5f}")
df_copy.head()
```

```python
df_copy = df.copy()

t0 = time.time()

df_copy["Sex"] = np.where(df_copy["Sex"]=="male", 1, 0)

t1 = time.time()
print(f"Processed in {t1-t0:.5f}")

df_copy.head()
```

```python
df_copy = df.copy()

t0 = time.time()

df_copy["Sex"] = df_copy["Sex"].replace("male", "1")
df_copy["Sex"] = df_copy["Sex"].replace("female", "0")
df_copy = df_copy.astype({"Sex": "int16"})

t1 = time.time()
print(f"Processed in {t1-t0:.5f}")

df_copy.head()
```

```python
df["Sex"] = np.where(df["Sex"]=="male", 1, 0)
df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

## Outliers

https://builtin.com/sites/www.builtin.com/files/styles/ckeditor_optimize/public/inline-images/1_boxplots.jpg

```python
df.boxplot(column=['Age'])
plt.show()
```

```python
q1 = np.quantile(df["Age"], 0.25)
q3 = np.quantile(df["Age"], 0.75)
print(f"Q1: {q1}, Q3: {q3}")

iqr = q3 - q1
print(f"IQR: {iqr}")

maximum = q3 + (1.5 * iqr)
minimum = q1 - (1.5 * iqr)
print(f"Min: {minimum}, Max: {maximum}")
```
