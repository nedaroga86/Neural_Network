```python
import statsmodels.api as sm
```

```python
x = df_temp['ApplicantIncome']
y = df_temp['LoanAmount']
 
# adding the constant term
x = sm.add_constant(x)
 
# performing the regression
# and fitting the model
model = sm.OLS(y, x).fit()

# printing the summary table
print(model.summary())
```

```python
df["Income"] = df["ApplicantIncome"] + df["CoapplicantIncome"]
df_temp["Income"] = df_temp["ApplicantIncome"] + df_temp["CoapplicantIncome"]
df.head()
```

```python
df_temp
```

```python
x = df_temp['Income']
y = df_temp['LoanAmount']
 
# adding the constant term
x = sm.add_constant(x)
 
# performing the regression
# and fitting the model
model = sm.OLS(y, x).fit()

# printing the summary table
print(model.summary())
```

```python
temp_test = df[df["LoanAmount"].isna()]["Income"]
temp_test = sm.add_constant(temp_test)
predictions = model.predict(temp_test)
print(len(predictions))
predictions
```

```python
import numpy as np
predictions = np.round(np.array(predictions))
predictions
```

```python
df.loc[df["LoanAmount"].isna(), "LoanAmount"] = predictions
df.isna().sum()
```
